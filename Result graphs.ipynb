```python
import matplotlib.pyplot as plt
```

```python
x_data = [640, 800, 1600, 1920, 3840]
y_seq_data = [6.278798, 9.805754, 29.638070, 43.966289, 177.867248]

y_openmp_data = [0.524497, 0.675236, 1.572185, 2.169725, 9.463065]

y_opencl_global_data = [0.116246, 0.171157, 0.460573, 0.648512, 2.518930]

y_opencl_local_data = [0.022038, 0.034357, 0.122467, 0.129836, 0.485366]
```

```python
plt.figure(figsize=(15, 10))
my_xticks = ['640:480', '800:600', '1600:900', '1920:1080', '3840:2160']
plt.xticks(y_data, my_yticks)

plt.plot(x_data, y_seq_data, label="seq")
plt.plot(x_data, y_openmp_data, label="openmp")
plt.plot(x_data, y_opencl_global_data, label="openCL global")
plt.plot(x_data, y_opencl_local_data, label="openCL local")

plt.legend()
plt.show()
```
